```python
import json
import pandas as pd

# Ruta al JSON que generaste
with open("booking_data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Convertir a DataFrame
df = pd.DataFrame(data)

# Ver primeras filas
```

```python
df.info()
```

```python
#Eliminar filas que no tienen precio
df = df[df["price"].str.replace(r"[^\d,]", "", regex=True).str.strip() != ""]
```

```python
print(df["price"].isna().sum())
```

```python
# Convertir price a n√∫mero (solo el valor num√©rico)
df["price_num"] = df["price"].str.replace(r"[^\d,]", "", regex=True).str.replace(",", ".").astype(float)
```

```python
# Convertir rating y pointService a float
df["rating"] = df["rating"].str.replace(",", ".").astype(float, errors="ignore")
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

df["pointService"] = df["pointService"].astype(float, errors="ignore")
```

```python
df["destino"] = df["destino"].str.lower().str.strip().astype("str")
```

```python
df.head(3)
```

```python
df.groupby("destino")["rating"].mean().round(2).sort_values(ascending=False)
```

```python
# Correcci√≥n bayesiana
df["weighted_score"] = (df["rating"] * (df["nreviews"] / (df["nreviews"] + 2))).round(2)
```

```python
df.head()
```

```python
def separar_resenas_por_sentimiento(reviews):
    positivos, negativos = [], []
    if isinstance(reviews, list):
        for r in reviews:
            pos = r.get("positive_feedback", "").strip()
            neg = r.get("negative_feedback", "").strip()
            if pos and pos.lower() not in ["n/a", "na", "no aplica"]:
                positivos.append(pos)
            if neg and neg.lower() not in ["n/a", "na", "no aplica"]:
                negativos.append(neg)
    return " ".join(positivos), " ".join(negativos)
```

```python
# Crear columnas separadas
df[["resenas_positivas", "resenas_negativas"]] = df["reviews"].apply(
    lambda x: pd.Series(separar_resenas_por_sentimiento(x))
)
```

```python

# Forzar conversi√≥n segura a float / int
df["weighted_score"] = pd.to_numeric(df["weighted_score"], errors="coerce")
df["pointService"] = pd.to_numeric(df["pointService"], errors="coerce")
df["price_num"] = pd.to_numeric(df["price_num"], errors="coerce")
df["nreviews"] = pd.to_numeric(df["nreviews"], errors="coerce")
```

```python
from collections import Counter
def contar_servicios(lista_de_listas):
    c = Counter()
    for servicios in lista_de_listas:
        if isinstance(servicios, list):
            c.update(servicios)
    return dict(c)

```

```python
destinos_grouped = (
    df.groupby("destino", dropna=False)
      .agg({
          "weighted_score": "mean",
          "pointService": "mean",
          "price_num": "mean",
          "nreviews": "sum",
          "resenas_positivas": lambda x: " ".join(filter(None, x)),
          "resenas_negativas": lambda x: " ".join(filter(None, x)),
          "title" : "count",
          "services": lambda x: contar_servicios(x),
      })
      .reset_index()
)
```

```python
#
destinos_grouped.rename(columns={"title": "n_hoteles"}, inplace=True)
```

```python
destinos_grouped["n_hoteles"] = pd.to_numeric(destinos_grouped["n_hoteles"], errors="coerce")
```

```python
def analizar_destino_gpt(nombre, positivos, negativos, score, point_service, n_hoteles, servicios_dict):
    positivos = (positivos or "")[:2500]
    negativos = (negativos or "")[:2500]

    servicios_text = ", ".join([f"{k}: {v}" for k, v in servicios_dict.items()]) if servicios_dict else "No disponible"

    prompt = f"""
    Analiza las rese√±as del destino **{nombre}** en Ecuador.

    Datos num√©ricos:
    - Puntaje corregido (weighted_score): {score}
    - Promedio de servicios (pointService): {point_service}
    - Hoteles analizados: {n_hoteles}

    Servicios disponibles (conteo por destino):
    {servicios_text}

    --- POSITIVOS ---
    {positivos}

    --- NEGATIVOS ---
    {negativos}

    Instrucciones:
    1. Resume los patrones m√°s repetidos en comentarios positivos y negativos.
    2. Da una nota de hospitalidad percibida (1-10) basada SOLO en las rese√±as. Si no hay base, di ‚Äúinsuficiente‚Äù.
    3. Eval√∫a la oferta tur√≠stica en funci√≥n de la variedad/frecuencia de servicios, sin asumir calidad.
    4. Perfil del destino (solo si las rese√±as lo permiten).
    5. Riesgos y atractivos para inversi√≥n tur√≠stica sin inventar nada.
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.3
    )
    return response.choices[0].message.content.strip()
```

```python
import time

def generar_analisis_por_destino(df):
    resultados = []
    for _, row in df.iterrows():
        try:
            print(f"Analizando destino: {row['destino']} ...")
            analisis = analizar_destino_gpt(
                nombre=row["destino"],
                positivos=row["resenas_positivas"],
                negativos=row["resenas_negativas"],
                score=row["weighted_score"],
                point_service=row["pointService"],
                n_hoteles=row["n_hoteles"],
                servicios_dict=row["services"]   # ‚Üê AQU√ç EL DICCIONARIO
            )
            resultados.append(analisis)
            time.sleep(1.5)
        except Exception as e:
            print(f"Error con {row['destino']}: {e}")
            resultados.append(None)

    df["analisis_gpt"] = resultados
    return df


destinos_grouped = generar_analisis_por_destino(destinos_grouped)
```
